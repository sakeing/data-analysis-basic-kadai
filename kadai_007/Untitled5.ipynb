```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib


df = pd.read_csv("sample_pandas_6.csv")
df.head()

cat_df = pd.read_csv('category.csv')
df = pd.merge(df,cat_df[['商品番号','カテゴリー']],how='inner', on='商品番号')

cat = df['カテゴリー'].value_counts()
cat.plot(kind='bar')
plt.show()

dfsum = df.groupby('商品番号')['注文数'].describe()
print(dfsum)
```
