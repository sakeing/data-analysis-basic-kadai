```python
import numpy as np
import matplotlib.pyplot as plt

nrows = 2
ncols = 3

fig,axes = plt.subplots(nrows,ncols,tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

x = np.linspace(0,10,100)
y = 2+2*np.sin(2*x)

axes[0,1].plot(x,y,linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))


hisx = np.random.normal(15,5,2000)
axes[0,2].hist(hisx)

np.random.seed(3)
scax = 4+np.random.normal(0,2,60)
scay = 4+np.random.normal(0,2,len(scax))

axes[1,0].scatter(scax,scay,s = 10)

plt.show()
```
